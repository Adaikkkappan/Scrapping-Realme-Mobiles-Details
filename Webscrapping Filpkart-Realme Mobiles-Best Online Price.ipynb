### Check the Status Code

```python
r=requests.get(url)

r.status_code
```

### Find all the tables used in the webpage

```python
all_tables=soup.find_all("table")
print(all_tables)
```

### Find the Model Class Name from the webpage and convert that into list

```python
model=soup.find_all('div',{'class':'_4rR01T'})
```

```python
model
```

```python
product_name=[]
for p in model:
    pdt=p.text
    product_name.append(pdt)
    
print(product_name)
```

### Find the Price Class Name from the webpage and convert that into list

```python
prices=soup.find_all('div',{'class':'_30jeq3 _1_WHN1'})
```

```python
price_list=[]
for p in prices:
    pl=p.text
    price_list.append(pl)
    
print(price_list)
```

### Find the Rating Class Name from the webpage and convert that into list

```python
ratings=soup.find_all('div',{'class':'_3LWZlK'})
ratings
```

```python
rating=[]
for r in ratings:
    ra=r.text
    rating.append(ra)
    
print(rating)
```

### Check the No. of values present in Rating, Product Name and Price List

### Convert the product_name,price_list,rating List into a Data Frame

```python
import pandas as pd
import numpy as np
d={'Product Name':product_name,'Prices':price_list,'Ratings':rating}
data=pd.DataFrame(d)
data
```

### Set an Index for a Data Frame
